```python
# data science
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
#load and combine csv files

df1 = pd.read_csv('resources/satisfaction_survey_20.csv')
df2 = pd.read_csv('resources/satisfaction_survey_80.csv')

df = pd.concat([df1,df2], ignore_index = True)
df.head()
```

```python
df.info()
```

```python
#changing Arrival Delay in Munites from float to int
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].replace([np.inf, -np.inf], np.nan)
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(0).astype(int)
df.info()
```

```python
df.describe()
```

```python
df.to_csv("satisfaction_data.csv", index=True, header=True)
```
